```python
import sqlite3
import os
from tabulate import tabulate

# Caminho para o banco
db_path = os.path.join("..", "unidades.db")

# Conectar ao banco
conn = sqlite3.connect(db_path)

# Substitua 'nome_da_tabela_correta' pelo nome correto da tabela
cursor = conn.cursor()
cursor.execute("SELECT DISTINCT estado_cdn FROM unidades WHERE estado_cdn IS NOT NULL;")
valores_unicos = cursor.fetchall()  # Retorna lista de tuplas

conn.close()

# Converter para lista simples
valores_unicos = [v[0] for v in valores_unicos]

# Exibir em tabela no console
tabela = [[i+1, estado] for i, estado in enumerate(sorted(valores_unicos))]
print(tabulate(tabela, headers=["#", "estado_cdn"], tablefmt="fancy_grid"))
```
